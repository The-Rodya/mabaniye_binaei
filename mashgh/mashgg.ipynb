```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def linear_gradient(width, height, point1, point2):
    img = np.zeros((height, width, 3), dtype=np.uint8)
    x1, y1, start_color = point1 ; x2, y2 , end_color= point2
    for x in range(width):
        for y in range(height):
            L = ((x - x1) * (x2 - x1) + (y - y1) * (y2 - y1)) / float((x2 - x1)**2 + (y2 - y1)**2)
            L = np.clip(L,0,1)
            img[y, x] = [int(start_color[0] * (1 - L) + end_color[0] * L),
                         int(start_color[1] * (1 - L) + end_color[1] * L),
                         int(start_color[2] * (1 - L) + end_color[2] * L)]

    return img
```

```python
width, height = 350, 350
# point = (x , y , color RGB)
point1 = (75,75,[200, 35, 103])
point2 = (350,350,[49, 130, 150])
grad = linear_gradient(width, height, point1, point2)
```

```python
plt.imshow(grad.astype(np.uint8))
plt.axis('off')
plt.show()


```
